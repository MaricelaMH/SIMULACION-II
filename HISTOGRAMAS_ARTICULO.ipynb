```python
import numpy as np
import matplotlib.pyplot as plt

# Datos de ejemplo
tiempos_espera = [18, 7, 12, 15, 16, 22, 21, 19, 27, 22, 27, 17]

# Crear el histograma
plt.figure(figsize=(6, 5))  # Tamao del grfico
plt.hist(tiempos_espera, bins=6, edgecolor='black', color='#4169E1', alpha=0.85)  # Color llamativo

# Agregar estadsticas clave
media = np.mean(tiempos_espera)
desviacion_estandar = np.std(tiempos_espera)
plt.axvline(x=media, color='red', linestyle='--', linewidth=1.5, label=f'Media = {media:.2f} min')
plt.axvline(x=media + desviacion_estandar, color='blue', linestyle='--', linewidth=1.5, label=f'+1 = {media + desviacion_estandar:.2f} min')
plt.axvline(x=media - desviacion_estandar, color='blue', linestyle='--', linewidth=1.5, label=f'-1 = {media - desviacion_estandar:.2f} min')

# Personalizacin del grfico
plt.title('Histograma de Tiempos de Espera', fontsize=14, fontweight='bold')
plt.xlabel('Tiempo de Espera (minutos)', fontsize=12)
plt.ylabel('Frecuencia', fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)  # Lneas de cuadrcula
plt.gca().set_facecolor('#F5F5F5')  # Fondo gris claro
plt.legend(fontsize=10)

# Mostrar el grfico
plt.tight_layout()
plt.show()

# Imprimir estadsticas
print(f"Media: {media:.2f} minutos")
print(f"Desviacin Estndar: {desviacion_estandar:.2f} minutos")

```

```python
import matplotlib.pyplot as plt

# Datos de tiempos de abordaje en minutos
tiempos_abordaje = [5,5,8,5,4,4,4,4,4,4,4,6,5]

# Crear el histograma
plt.hist(tiempos_abordaje, bins=6, edgecolor='black', color='#BA55D3', alpha=0.85)  # Color llamativo

# Agregar lneas de referencia
promedio = sum(tiempos_abordaje) / len(tiempos_abordaje)
plt.axvline(x=promedio, color='red', linestyle='--', label=f'Promedio ({promedio:.2f} min)')

# Personalizacin del grfico
plt.title('Histograma de Tiempos de Abordaje', fontsize=14, fontweight='bold')
plt.xlabel('Tiempo de Abordaje (minutos)', fontsize=12)
plt.ylabel('Frecuencia', fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)  # Lneas de cuadrcula
plt.gca().set_facecolor('#F5F5F5')  # Fondo gris claro
plt.legend(fontsize=10)

# Mostrar el grfico
plt.tight_layout()
plt.show()

```

```python
import matplotlib.pyplot as plt

#  Capacidad mxima y pasajeros abordados por combi
capacidad_maxima = [19] * 13  # Todas las combis tienen capacidad de 12 pasajeros
pasajeros_abordados = [18, 18, 19, 19, 18, 19, 19, 19, 19, 19, 19, 19,19]  # Pasajeros reales por combi

# Calcular porcentaje de ocupacin
porcentaje_ocupacion = [(abordados / capacidad) * 100 for abordados, capacidad in zip(pasajeros_abordados, capacidad_maxima)]

# Crear el histograma
plt.hist(porcentaje_ocupacion, bins=2, edgecolor='black', color='#008080', alpha=0.7)

# Personalizar el grfico
plt.title('Histograma de Porcentaje de Ocupacin')
plt.xlabel('Porcentaje de Ocupacin (%)', fontsize=12)
plt.ylabel('Frecuencia', fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)  # Lneas de cuadrcula
plt.gca().set_facecolor('#F5F5F5')  # Fondo gris claro
plt.axvline(x=100, color='red', linestyle='--', label='Capacidad completa (100%)')
plt.legend(fontsize=10)

# Mostrar el grfico
plt.tight_layout()
plt.show()


```

```python
import matplotlib.pyplot as plt

# Datos: tasas de llegada y servicio
tasas_llegada = [0.62,0.95,1.27]  # Lambda (por minuto)
tasas_servicio = [0.25,0.25,0.21]  # Mu (por minuto)

# Calcular la tasa de ocupacin (rho)
tasa_ocupacion = [l / m for l, m in zip(tasas_llegada, tasas_servicio)]

# Crear el histograma
plt.hist(tasa_ocupacion, bins=6, edgecolor='black', color='#FF8C00', alpha=0.7)

# Lnea de referencia para la estabilidad
plt.axvline(x=1, color='red', linestyle='--', label='Estabilidad lmite (=1)')

# Personalizar el grfico
plt.title('Histograma de la Tasa de Ocupacin ()')
plt.xlabel('Tasa de Ocupacin ()')
plt.ylabel('Frecuencia', fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)  # Lneas de cuadrcula
plt.gca().set_facecolor('#F5F5F5')  # Fondo gris claro
plt.legend(fontsize=10)

# Mostrar el grfico
plt.tight_layout()
plt.show()
```
