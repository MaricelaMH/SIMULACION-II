```python
import yfinance as yf
import matplotlib.pyplot as plt
import numpy as np
```

```python
tesla = yf.Ticker("TSLA")
df = tesla.history(start="2023-12-13",interval="1d")
#df.head(10)
```

```python
tesla_cierre=df["Close"]
#tesla_cierre
```

```python
amazon = yf.Ticker("AMZN")
df = amazon.history(start="2023-12-13",interval="1d")
#df.head(10)
```

```python
amazon_cierre=df["Close"]
#amazon_cierre
```

```python
uber = yf.Ticker("UBER")
df = uber.history(start="2023-12-13",interval="1d")
#df.head(10)
```

```python
uber_cierre=df["Close"]
#uber_cierre
```

```python
nestle = yf.Ticker("NESN.SW")
df = nestle.history(start="2023-12-13",interval="1d")
#df.head(10)
```

```python
nestle_cierre=df["Close"]
#bimbo_cierre
```

```python
plt.figure(figsize=(10, 6))
plt.plot(tesla_cierre, color='blue', label="Tesla")
plt.plot(amazon_cierre, color='orange', label="Amazon")
plt.plot(uber_cierre, color='green', label="Uber")
plt.plot(nestle_cierre, color='red', label="Nestlé")
plt.title("Cierres históricos")
plt.xlabel("Fecha")
plt.ylabel("Precio")
plt.legend()
plt.grid(True)
plt.show()
```

Pero es mas util trabajar con rendimientos o retornos

$$P = \frac{S_i - S_{i-1}}{S_i}  $$

```python
tesla_retornos = np.log(tesla_cierre)-np.log(tesla_cierre.shift(1))
tesla_ret = tesla_retornos.dropna()
```

```python
amazon_retornos = np.log(amazon_cierre)-np.log(amazon_cierre.shift(1))
amazon_ret = amazon_retornos.dropna()
```

```python
uber_retornos = np.log(uber_cierre)-np.log(uber_cierre.shift(1))
uber_ret = uber_retornos.dropna()
```

```python
nestle_retornos = np.log(nestle_cierre)-np.log(nestle_cierre.shift(1))
nestle_ret = nestle_retornos.dropna()
```

```python
plt.figure(figsize=(10, 6))
plt.plot(tesla_ret, color='blue', label="Tesla")
plt.plot(amazon_ret, color='orange', label="Amazon")
plt.plot(uber_ret, color='green', label="Uber")
plt.plot(nestle_ret, color='red', label="Nestlé")
plt.title("Retornos")
plt.xlabel("Fecha")
plt.ylabel("Precio")
plt.legend()
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(12, 6))
plt.hist(tesla_ret, bins=30, alpha=0.5, label='Tesla', color='blue')
plt.hist(amazon_ret, bins=30, alpha=0.5, label='Amazon', color='orange')
plt.hist(uber_ret, bins=30, alpha=0.5, label='Uber', color='green')
plt.hist(nestle_ret, bins=30, alpha=0.5, label='Nestle', color='red')
plt.title('Histograma de Retornos')
plt.xlabel('Retornos')
plt.ylabel('Frecuencia')
plt.legend()
plt.grid(True)
plt.show()
```
