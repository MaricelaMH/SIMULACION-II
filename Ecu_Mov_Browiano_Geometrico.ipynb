```python
import random as rd
import matplotlib.pyplot as plt
```

```python
def caminante(N, p):
    p = p
    N = N
    Cont = 0
    L = [Cont]
    for i in range(N):
     t = rd.uniform(0,1)
     if t < p:
        Cont += 1
     else:
        Cont -= 1
     L.append(Cont)
    plt.plot(L)
```

```python
for i in range(100):
 caminante(400,0.6)
```

```python
for i in range(100):
 caminante(400,0.5)
```

```python
for i in range(100):
 caminante(400,0.3)
```

NOTA: Cuando la probabiliodad es menor a 0.5 , la tendencia va hacia abajo, y cuando la probabilidad es mayor a 0.5, la tendencia va hacia arriba.
