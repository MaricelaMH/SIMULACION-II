```python
import sympy as sp
sp.init_printing()
```

```python
# Grafica del seno
t=sp.symbols('t')
sp.plotting.plot(sp.sin(t))
```

```python
#Funcion escalon
Y = sp.Heaviside(t-sp.pi)
sp.plotting.plot(Y)
```

```python
u = t**2+(t-t**2)*Y
sp.plotting.plot(u)
```

```python
#Funcion escalon
E1 = (1/2*sp.Heaviside(t-2))-(1/2*sp.Heaviside(t-4))
sp.plotting.plot(E1,(t,0,6))
```

```python
#
E2 = (1*sp.Heaviside(t-2.5))-(1*sp.Heaviside(t-3.5))
sp.plotting.plot(E2,(t,0,6))
```

```python
# Funcion delta dirac
D = sp.DiracDelta(t)
sp.plotting.plot(D)
```
