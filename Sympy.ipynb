```python
import sympy as sp
sp.init_printing()
```

1. Definimos los simbolos (variables a usar S,c t)

```python
S=sp.symbols('S')
c=sp.symbols('c')
t=sp.symbols('t')
```

2. Observamos la ecuación definida, con expand() realiza el calculo simbolico

```python
((t+4)**2).expand()
```

3. Con subs(variable a remplazar,valor), resuelve la ecuación

```python
((t+4)**2).subs(t,3)
```

4. Con factor() factorizamos la ecuación

```python
sp.factor((t**2)+6*t+5)
```

5. Con simplify() simplificamos la funcion

```python
sp.simplify((t**2)+6*t+5+(t**2))
```

6. Con sp.plotting.plot(funcion) grafica nuestra funcion

```python
sp.plotting.plot(t*t)
```

```python
sp.plotting.plot(5*sp.exp(0.5*t)) # 0<c<1
```

7.Definimos nuestra EDO

```python
S=sp.Function('S')
ode=sp.Eq(sp.Derivative(S(t),t)-c*S(t),0)
ode
```

8. Obtenemos la solucion de la EDO, dandole ya valor a nuestra constante A, ics= son nuestras condiciones

```python
sol = sp.dsolve(ode,S(t),ics={S(0):10000})
sol
```

```python
constants = {c:0.05}
sol.subs(constants)
```

```python
sp.plotting.plot(sol.rhs.subs(constants))
```
